## Matrix of Transitions

```python
graph = np.zeros((8, 8))
graph[0, 1:3] = 1
graph[1, (0, 2, 3)] = 1
graph[2, (0, 1, 3)] = 1
graph[3, (1, 2, 4)] = 1
graph[4, (3, 5, 6)] = 1
graph[5, (4, 6, 7)] = 1
graph[6, (4, 5, 7)] = 1
graph[7, (5, 6)] = 1

# graph[0, 2] = 1
# graph[2, 0] = 1

# graph[1, 2] = 1
```

```python
G = nx.Graph() 
for i in range(graph.shape[0]):
    for j in range(graph.shape[1]):
        if graph[i][j]:
            G.add_edge(i, j)
```

```python
nx.draw(G, with_labels = True)
```

## Generate different paths

```python
#Random Walk
start_state = np.random.choice(range(8))
path_length = 10000
current_state = start_state
path = np.zeros(path_length)
for i in range(path_length):
    path[i] = current_state
    neighbour_states = np.where(graph[current_state])[0]
    next_state = np.random.choice(neighbour_states)
    current_state = next_state
    
```

## SR Learning

```python
def draw_SR_categories(cutoff_point, alpha, gamma):
    SR = np.random.uniform(0, 1, size=(8, 8))
    start_state = np.random.choice(np.arange(8))
    current_state = start_state

    for i, observed_state in enumerate(path[:cutoff_point]):

        expected_probs = SR[current_state]
        one_hot_obs = np.zeros(8)
        one_hot_obs[int(observed_state)] = 1

        SR_delta = one_hot_obs + gamma*SR[int(observed_state), :] - expected_probs

        SR[current_state, :] = SR[current_state, :] + alpha*SR_delta
        SR[current_state, :] = SR[current_state, :]/sum(SR[current_state])
        current_state = int(observed_state)    


    G = nx.Graph() 
    for i in range(SR.shape[0]):
        for j in range(SR.shape[1]):
            G.add_edge(i, j, weight = SR[i][j])    
    edges = G.edges
    weights = [SR[u][v] for u,v in edges]


    kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(SR)
    node_color = []

    for node in range(8):
        if kmeans.labels_[node]:
            node_color.append('red')
        else:
            node_color.append('blue')

    nx.draw(G, width = weights, with_labels = True, node_color = node_color)
    plt.show()

```

```python
cutoff_point = widgets.IntSlider(
    value=7,
    min=0,
    max=10000,
    step=1,
    description='Cutoff Point',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
```

```python
draw_SR_categories(30, 0.1, 0.8)
```
